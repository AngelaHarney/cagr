```python
import subprocess
import os

# Step 1: Check if snow CLI is available
result = subprocess.run(['which', 'snow'], capture_output=True, text=True)
print(f'snow location: {result.stdout.strip()}')
print(f'snow which error: {result.stderr.strip()}')

# Step 2: Check snow version
result = subprocess.run(['snow', '--version'], capture_output=True, text=True)
print(f'snow version: {result.stdout.strip()}')
print(f'snow version error: {result.stderr.strip()}')
```

```python
import subprocess
import os

# Step 3: If snow is available, try running dcm plan
# First, install snowflake-cli if not present
result = subprocess.run(['pip', 'install', 'snowflake-cli'], capture_output=True, text=True)
print(f'pip install: {result.returncode}')
print(result.stderr[-500:] if result.stderr else 'no errors')
```

```python
import subprocess
import os

# Step 4: Try the actual DCM plan command
# This is the critical test - can it upload to the internal stage?
os.chdir('/workspace/dcm')
result = subprocess.run(
    ['snow', 'dcm', 'plan', 'PRECISION_DEV.CAGR.PRECISION_CALCS', '-c', 'default', '--target', 'DEV'],
    capture_output=True, text=True, timeout=120
)
print(f'returncode: {result.returncode}')
print(f'stdout: {result.stdout[:2000]}')
print(f'stderr: {result.stderr[:2000]}')
```
